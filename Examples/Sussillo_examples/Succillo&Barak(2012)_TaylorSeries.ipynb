```python
# Taylor series of f(x) = e^x  (n = 3)
# -----------------------------------

# Taylor series 
# x = sympy symbol
# x0 = intercept
# n = how many factors to go
x = sympy.Symbol('x')
taylor = sympy.series(sympy.exp(x), x = x, x0 = 0, n = 3)
print(taylor)
```
